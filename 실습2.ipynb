```python
import csv

f = open('subwayfee2.csv')
data = csv.reader(f)
next(data)

temp = []
for row in data:
    temp.append(row)

top = [0, 0, 0]
mx_station = ['', '', '']
station_id = ['', '', '']

for i in range(0, 3):
    index = 0
    delete_index = 0
    for row in temp: 
        if temp[index] is None:
            continue
        users = int(row[4].replace(",", "")) + int(row[5].replace(",", ""))
        if top[i] < users:
            top[i] = users
            mx_station[i] = row[1]
            station_id[i] = row[3]
            delete_index = index

        index = index + 1
    temp[delete_index] = None
    
for i in range(0, 3):
    print(mx_station[i])
    print(station_id[i])
    print(top[i])
    print("\n")   
    
f.close()  


f = open('subwayfee2.csv')
data = csv.reader(f)
next(data)

temp = []
for row in data:
    temp.append(row)
    
below = [999, 999, 999]
mn_station = ['', '', '']
station_id2 = ['', '', '']

for i in range(0, 3):    
    index = 0
    delete_index = 0
    for row in temp:
        if temp[index] is None:
            continue
        users = int(row[4].replace(",", "")) + int(row[5].replace(",", ""))
        if below[i] > users:
            below[i] = users
            mn_station[i] = row[1]
            station_id2[i] = row[3]
            delete_index = index

        index = index + 1
    temp[delete_index] = None
    
for i in range(0, 3):
    print(mn_station[i])
    print(station_id2[i])
    print(below[i])
    print("\n")
    
f.close()  
```
