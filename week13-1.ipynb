```python
import csv
import matplotlib.pyplot as plt

f = open('seoul1.csv')
data = csv.reader(f)
next(data)
high_result=[]
low_result=[]

        
for row in data :
    high_temp=row[4]
    low_temp=row[5]
    date=row[2]
    if date=='' :
        break
    year,month,day = date.split('-')[:3]
    
    if row[-1] != '' :
        if month == '10' and day == '17' :
            high_result.append(float(high_temp))
            low_result.append(float(low_temp))

plt.plot(high_result)
plt.plot(low_result)
plt.show()
```
